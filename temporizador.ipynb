```python
import tkinter as tk

import random

import time

window = tk.Tk()

window.title("Temporizador o cronometro")

canvas = tk.Canvas(window, width=500, height=300)

canvas.pack()

opcion_seleccionada = tk.StringVar()
opcion_seleccionada.set("None")

segundos = 0
minutos = 0
horas = 0

timegundos = 0
timinutos = 0
timehoras = 0

tiempo = tk.Label(canvas, text="00:00:00", width=10, height=3, font=("Helvetica", 56))

tiempo.place(x=248, y=88, anchor="center")

def iniciar():
    global opcion_seleccionada
    global timegundos
    global timinutos
    global timehoras
    global tiempo
    if opcion_seleccionada.get() == "Cronometro":
        while timegundos <= 10:
            time.sleep(1)
            if timegundos == 60:
                timegundos = 0
                timinutos += 1
            if timinutos == 60:
                timehoras += 1
                timinutos = 0
            timegundos += 1

            if timegundos < 10:
                tiempo.config(text=f"00:00:0{timegundos}")
                if timehoras < 10:
                    tiempo = tk.Label(canvas, text=f"0{timehoras}:00:{timegundos}", width=10, height=3, font=("Helvetica", 56))
                    tiempo.place(x=248, y=88, anchor="center")
                    if timinutos < 10:
                        tiempo = tk.Label(canvas, text=f"0{timehoras}:0{timinutos}:0{timegundos}", width=10, height=3, font=("Helvetica", 56))
                        tiempo.place(x=248, y=88, anchor="center")
                if timinutos < 10:
                    tiempo = tk.Label(canvas, text=f"00:0{timinutos}:0{timegundos}", width=10, height=3, font=("Helvetica", 56))
                    tiempo.place(x=248, y=88, anchor="center")
            if timinutos < 10:
                tiempo = tk.Label(canvas, text=f"00:0{timinutos}:00", width=10, height=3, font=("Helvetica", 56))
                tiempo.place(x=248, y=88, anchor="center")
                if timehoras < 10:
                    tiempo = tk.Label(canvas, text=f"0{timehoras}:0{timinutos}:00", width=10, height=3, font=("Helvetica", 56))
                    tiempo.place(x=248, y=88, anchor="center")
                    if timegundos < 10:
                        tiempo = tk.Label(canvas, text=f"0{timehoras}:0{timinutos}:0{timegundos}", width=10, height=3, font=("Helvetica", 56))
                        tiempo.place(x=248, y=88, anchor="center")
                if timegundos < 10:
                    tiempo = tk.Label(canvas, text=f"00:0{timinutos}:0{timegundos}", width=10, height=3, font=("Helvetica", 56))
                    tiempo.place(x=248, y=88, anchor="center")
            if timehoras < 10:
                tiempo = tk.Label(canvas, text=f"0{timehoras}:00:00", width=10, height=3, font=("Helvetica", 56))
                tiempo.place(x=248, y=88, anchor="center")
                if timinutos < 10:
                    tiempo = tk.Label(canvas, text=f"0{timehoras}:0{timinutos}:00", width=10, height=3, font=("Helvetica", 56))
                    tiempo.place(x=248, y=88, anchor="center")
                    if timegundos < 10:
                        tiempo = tk.Label(canvas, text=f"0{timehoras}:0{timinutos}:0{timegundos}", width=10, height=3, font=("Helvetica", 56))
                        tiempo.place(x=248, y=88, anchor="center")
                if timegundos < 10:
                    tiempo = tk.Label(canvas, text=f"0{timehoras}:00:0{timegundos}", width=10, height=3, font=("Helvetica", 56))
                    tiempo.place(x=248, y=88, anchor="center")
            
            

boton = tk.Button(canvas, text="Iniciar", command=iniciar, width=6, height=1)

boton.place(x=150, y=248, anchor="center")
# Valor predeterminado

def check():
    global opcion_seleccionada
    global segundos
    global minutos
    global horas
    if opcion_seleccionada.get() == "Temporizador":
        print("Temporizador")
        segundos = tk.Entry(canvas, width=5)
        segundos.place(x=370, y=130)
        minutos = tk.Entry(canvas, width=5)
        minutos.place(x=370, y=150)
        horas = tk.Entry(canvas, width=5)
        horas.place(x=370, y=170)
    if opcion_seleccionada.get() == "Cronometro":
        if segundos != 0:
            segundos.place(x=1000, y=1000)
        if minutos != 0:
            minutos.place(x=1000, y=1000)
        if horas != 0:
            horas.place(x=1000, y=1000)
        print("Cronometro")


# Crear Radiobuttons
radiobutton1 = tk.Radiobutton(window, text="Temporizador", variable=opcion_seleccionada, value="Temporizador", command=check)

radiobutton2 = tk.Radiobutton(window, text="Cronometro", variable=opcion_seleccionada, value="Cronometro", command=check)

radiobutton1.place(x=45, y=158)

radiobutton2.place(x=45, y=188)

window.mainloop()
```
