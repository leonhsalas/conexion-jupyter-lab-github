```python
import tkinter as tk


window = tk.Tk()

window.title("Juego del ahorcado")

canvas = tk.Canvas(window, width=400, height=210)

canvas.pack()

numero = 0

def abrir():
    nueva_ventana = tk.Toplevel(window)
    nueva_ventana.title("Juego del ahorcado")
    canvas2 = tk.Canvas(nueva_ventana, width=400, height=210)
    canvas2.pack()
    palabra = tk.Entry(canvas2, width=7)

    def comenzar():
        if palabra.get() == "":
            print("No tienes una palabra")
        else:
            pal = (palabra.get()).lower()
            contador = 0
            if len(pal) >= 9:
                print("Pon una palabra con m√°ximo de ocho caracteres")
            else:
                man = False
                while contador < len(pal):
                    if pal[contador] != "a" and pal[contador] != "b" and pal[contador] != "c" and pal[contador] != "d" and pal[contador] != "e" and pal[contador] != "f" and pal[contador] != "g" and pal[contador] != "h" and pal[contador] != "i" and pal[contador] != "j" and pal[contador] != "k" and pal[contador] != "l" and pal[contador] != "m" and pal[contador] != "n" and pal[contador] != "o" and pal[contador] != "p" and pal[contador] != "q" and pal[contador] != "r" and pal[contador] != "s" and pal[contador] != "t" and pal[contador] != "u" and pal[contador] != "v" and pal[contador] != "w" and pal[contador] != "x" and pal[contador] != "y" and pal[contador] != "z":
                        man = True
                        break
                    contador += 1
                if man == False:
                    juego = tk.Toplevel(window)
                    juego.title("Juego del ahorcado")
                    canvas3 = tk.Canvas(juego, width=600, height=300)
                    canvas3.pack()
    
                    maquin1 = canvas3.create_line(50, 250, 50, 50, fill="black", width=2)
                    maquin2 = canvas3.create_line(50, 50, 150, 50, fill="black", width=2)
                    maquin3 = canvas3.create_line(150, 50, 150, 70, fill="black", width=2)
    
                    contador = 0
    
                    x = 100
    
                    while contador < len(pal):
                        canvas3.create_line(x, 250, x+40, 250, fill="black", width=2)
                        contador += 1
    
                        x += 50
    
                    equivocado = tk.Listbox(canvas3, width=15, height=6)
    
                    equivocado.place(x=448, y=38)
    
                    def check(event):
                        global numero
                        if (event.keysym).lower() == pal:
                            print(pal)
                        else:
                            equivocado.insert(numero, (event.keysym).lower())
                            numero += 1
    
                    juego.bind('<Key>', check)
                    
                    juego.mainloop()
                else:
                    print("Pon una palabra con letras solamente sin espacios")
        
    Comenzar = tk.Button(canvas2, text="Comenzar", command=comenzar, width=6, height=1)
        
    Comenzar.place(x=280, y=150)
        
    palabra.place(x=200, y=105, anchor="center")
    nueva_ventana.mainloop()

Jugar = tk.Button(canvas, text="Jugar", command=abrir, width=6, height=1)

Jugar.place(x=160, y=95)

window.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk

class HorizontalListbox(tk.Frame):
    def __init__(self, parent, data, *args, **kwargs):
        super().__init__(parent, *args, **kwargs)

        self.canvas = tk.Canvas(self, borderwidth=0, background="#ffffff")
        self.frame = tk.Frame(self.canvas, background="#ffffff")
        self.vsb = tk.Scrollbar(self, orient="horizontal", command=self.canvas.xview)
        self.canvas.configure(xscrollcommand=self.vsb.set)

        self.vsb.pack(side="bottom", fill="x")
        self.canvas.pack(side="left", fill="both", expand=True)
        self.canvas.create_window((4, 4), window=self.frame, anchor="nw",
                                  tags="self.frame")

        self.frame.bind("<Configure>", self.onFrameConfigure)

        self.populate(data)

    def populate(self, data):
        for index, item in enumerate(data):
            label = tk.Label(self.frame, text=item, padx=10, pady=10)
            label.grid(row=0, column=index, sticky="nsew")

    def onFrameConfigure(self, event):
        self.canvas.configure(scrollregion=self.canvas.bbox("all"))

if __name__ == "__main__":
    root = tk.Tk()
    root.title("Horizontal Listbox Example")

    data = ["6", "7", "8", "9", "p", "8"]
    listbox = HorizontalListbox(root, data)
    listbox.pack(side="top", fill="both", expand=True)

    root.mainloop()
```
